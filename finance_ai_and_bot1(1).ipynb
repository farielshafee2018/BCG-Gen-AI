```python
company = {"APPLE","TESLA","MICROSOFT"}
```

```python
year ={"2022","2023","2024","2025"}
```

```python
def fetch_answers(user_query):
    for comp in company:
        if comp in str(user_query).upper():
            find_data(comp,user_query)
            break
        
    general_queries(user_query)

```

```python
def find_data(comp, user_query):
    for yr in year:
        if yr in str(user_query):
            comp_yrly(comp, yr, user_query)
            break
        
    comp_mean(comp, user_query)
```

```python
def comp_mean(comp, user_query):
    if "revenue" in str(user_query).lower():
        revmean=df.groupby("COMPANY")["TOT_REV_MIL"].mean().sort_values(ascending=False)[comp]
        print(f" Mean revenue for {comp} is {revmean}")
    elif "liability" in str(user_query).lower():
        lialmean= df.groupby("COMPANY")["TOT_LIAL_MIL"].mean().sort_values(ascending=False)[comp]
        print(f" Mean revenue for {comp} is {lialmean}")

    elif "income" in str(user_query).lower():
        incmean= df.groupby("COMPANY")["NET_INC_MIL"].mean().sort_values(ascending=False)[comp]
        print(f" Mean revenue for {comp} is {incmean}")
    elif "asset" in str(user_query).lower():
        assmean= df.groupby("COMPANY")["TOT_ASS_MIL"].mean().sort_values(ascending=False)[comp]
        print(f" Mean revenue for {comp} is {assmean}")
    elif "cash flow" in str(user_query).lower():
        cashmean= df.groupby("COMPANY")["CASH_FLOW_MIL"].mean().sort_values(ascending=False)[comp]
        print(f" Mean revenue for {comp} is {cashmean}")

    else:
        print("I do not know the answer. Please ask something else.")
        
        
```

```python
quit_words={"quit", "exit","good bye","bye"}
```

```python
def general_queries(user_query):
    for wrd in quit_words:
        if wrd in str(user_query):
            break
    else:
        print("Good day.  Kindly ask me something specific.  Thank you.")
```

```python
def comp_yrly(comp,yr, user_query):
     print(f"The year you have chosen is {yr} and the company you have chosen is {comp}")
     if "revenue growth" in str(user_query).lower():
        revyr= df[(df["COMPANY"]==comp) & (df["YEAR"]==int(yr))]["REV_GROWTH%"]
        print(f" revenue for {comp} is {revyr}")
     elif "income growth" in str(user_query).lower():
        incgrowth= df[(df["COMPANY"]==comp) & (df["YEAR"]==int(yr))]["INC_GROWTH%"]
        print(f" income growth for {comp} is {incgrowth}")
     elif "revenue" in str(user_query).lower():
        revgrowth= df[(df["COMPANY"]==comp) & (df["YEAR"]==int(yr))]["TOT_REV_MIL"]
        print(f" income growth for {comp} is {revgrowth}")
     elif "liability" in str(user_query).lower():
         lialyr=df[(df["COMPANY"]==comp) & (df["YEAR"]==int(yr))]["TOT_LIAL_MIL"]
         print(f" LIABILITY for {comp} is {lialyr}")
     elif "income" in str(user_query).lower():
        incyr= df[(df["COMPANY"]==comp) & (df["YEAR"]==int(yr))]["NET_INC_MIL"]
        print(f" INCOME for {comp} is {incyr}")
     elif "asset" in str(user_query).lower():
        assyr= df[(df["COMPANY"]==comp) & (df["YEAR"]==int(yr))]["TOT_ASS_MIL"]
        print(f" ASSET for {comp} is {assyr}")
     elif "cash flow" in str(user_query).lower():
        cashyr= df[(df["COMPANY"]==comp) & (df["YEAR"]==int(yr))]["CASH_FLOW_MIL"]
        print(f" CASH FLOW for {comp} is {cashyr}")
  
     else:
        print("I do not know the answer. Please ask something else.")
        
    
```

```python
def bot_prompt_loop():
    """
    Continuously prompts the user for input.
    """
    print("Welcome to the financial chatbot")
    print("Type 'exit' (or 'EXIT', 'Exit', etc.) to quit at any time.")

    while True:
        # Get user input
        user_query = input("Ask your question: ")

        # Check if the standardized input matches the exit command
        if user_query.lower() == 'exit':
            print("Exiting bot. Goodbye!")
            break
        
        # If the user input is not 'exit', process or display it
        if user_query:
            print(f"You entered: {user_query}")
            fetch_answers(user_query)
        else:
            print("Input cannot be empty. Please enter a string or 'exit'.")

if __name__ == '__main__':
    bot_prompt_loop()
    
```
